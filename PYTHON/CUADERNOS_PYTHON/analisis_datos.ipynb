```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Crear un DataFrame con datos de ventas mensuales
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'],
    'Ventas': [12000, 15000, 13500, 17000, 16000, 18000]
}

df = pd.DataFrame(datos)
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Mes'], df['Ventas'], marker='o', color='blue', linestyle='-')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Promedio de ventas:", df['Ventas'].mean())
print("Máximo de ventas:", df['Ventas'].max())
print("Mes con mayores ventas:", df.loc[df['Ventas'].idxmax(), 'Mes'])
```

```python
plt.bar(df['Mes'], df['Ventas'], color='skyblue')
plt.title('Ventas Mensuales (Gráfico de Barras)')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.show()
```
